We will create new features and perform data aggregation to prepare for further analysis or modeling.

```python
import pandas as pd

# Define the file path 
file_path = {
    'orders_train' : '/Users/dhari/OneDrive/Documentos/Work/Projects/Ecommerce project/Ecommerce Order Dataset/train/df_Orders.csv',
    'customers_train' : '/Users/dhari/OneDrive/Documentos/Work/Projects/Ecommerce project/Ecommerce Order Dataset/train/df_Customers.csv',
    'products_train' : '/Users/dhari/OneDrive/Documentos/Work/Projects/Ecommerce project/Ecommerce Order Dataset/train/df_Products.csv',
    'payments_train' : '/Users/dhari/OneDrive/Documentos/Work/Projects/Ecommerce project/Ecommerce Order Dataset/train/df_Payments.csv',
    'order_items_train' : '/Users/dhari/OneDrive/Documentos/Work/Projects/Ecommerce project/Ecommerce Order Dataset/train/df_OrderItems.csv',
}

# Load datasets
df_orders_train = pd.read_csv(file_paths['orders_train'])
df_customers_train = pd.read_csv(file_paths['customers_train'])
df_products_train = pd.read_csv(file_paths['products_train'])
df_payments_train = pd.read_csv(file_paths['payments_train'])
df_order_items_train = pd.read_csv(file_paths['order_items_train'])
```

```python
# Merge order items with orders to bring in customer_id
df_order_items_train = pd.merge(df_order_items_train, df_orders_train[['order_id', 'customer_id']], on='order_id', how='left')

# Merge with products and payments
df_order_items_train = pd.merge(df_order_items_train, df_products_train, on='product_id', how='left')
df_order_items_train = pd.merge(df_order_items_train, df_payments_train, on='order_id', how='left')
```

```python
# Calculate total order value
df_order_items_train['Total_Value'] = df_order_items_train['price'] + df_order_items_train['shipping_charges']

# Aggregating data by customer
customer_aggregates = df_order_items_train.groupby('customer_id').agg({
    'Total_Value': 'sum',
    'order_id': 'count'
}).reset_index()
customer_aggregates.columns = ['Customer_ID', 'Total_Spend', 'Total_Orders']

# Display the first few rows
display(customer_aggregates.head())
```
